```python
print("Tip Calculator")
print()

def valid_input(get_input):
    while True:
        try:
            value = float(input(get_input))
            if value <= 0:
                print("Must be greater than 0. Please try again.")
            else:
                return value
        except ValueError:
            print("Must be valid decimal number. Please try again.")

def valid_tip(get_input):
    while True:
        try:
            value = int(input(get_input))
            if value < 0 or value > 100:
                print("Must be valid integer between 0 and 100. Please try again.")
            else:
                return value
        except ValueError:
            print("Must be valid integer. Please try again.")

Cost = valid_input("Cost of meal: ")

print()
tip_percent = valid_tip("Tip percent:\t")
tip = round(Cost * (tip_percent / 100), 2)
total = round(Cost * (1 + tip_percent / 100), 2)

print("\nOUTPUT")
print("Cost of meal:\t", format(Cost, '.2f'))
print("Tip percent:\t", tip_percent, "%")
print("Tip amount:\t", format(tip, '.2f'))
print("Total amount:\t", format(total, '.2f'))






```

```python
import random

ALL_ITEMS_FILE = "wizard_all_items.txt"
INVENTORY_FILE = "wizard_inventory.txt"

def load_all_items():
    try:
        with open(ALL_ITEMS_FILE, "r") as file:
            return [line.strip() for line in file]
    except FileNotFoundError:
        print("Error: The all items file is missing.")
        exit()

def load_inventory():
    try:
        with open(INVENTORY_FILE, "r") as file:
            return [line.strip() for line in file]
    except FileNotFoundError:
        print("Warning: The inventory file is missing. Starting with an empty inventory.")
        return []

def save_inventory(items):
    with open(INVENTORY_FILE, "w") as file:
        for item in items:
            file.write(item + "\n")

def display_items(items):
    for i in range(len(items)):
        print(str(i + 1) + ". " + items[i])
    print()

def walk(all_items, current_inventory):
    if len(current_inventory) >= 4:
        print("You can't carry any more items. Drop something first.")
        return current_inventory

    item_to_add = random.choice(all_items)
    print("While walking down a path, you see " + item_to_add + ".")
    
    grab = input("Do you want to grab it? (y/n): ").lower()
    while grab not in ['y', 'n']:
        print("Invalid input. Please enter 'y' or 'n'.")
        grab = input("Do you want to grab it? (y/n): ").lower()  
        
    if grab == "y":
        if len(current_inventory) >= 4:
            print("You can't carry any more items. Drop something first.")
        else:
            current_inventory.append(item_to_add)
            print("You picked up " + item_to_add + ".")
            print()
            
    return current_inventory


def drop(current_inventory):
    if not current_inventory:
        print("Inventory is empty.")
        return current_inventory

    print("Inventory:")
    display_items(current_inventory)
    try:
        index = int(input("Drop Number: ")) - 1
        if index < 0 or index >= len(current_inventory):
            print("Invalid item number.")
        else:
            dropped_item = current_inventory.pop(index)
            print("You dropped " + dropped_item +".")
            print()
    except ValueError:
        print("Invalid input. Please enter a number.")
    return current_inventory

def main():
    all_items = load_all_items()
    current_inventory = load_inventory()

    print("The Wizard Inventory Program")
    print()
    print("COMMAND MENU")
    print("walk - Walk down the path")
    print("show - Show all items")
    print("drop - Drop an item")
    print("exit - Exit program")
    print()
    while True:
        command = input("Command: ").lower()
        print()
        if command == "walk":
            current_inventory = walk(all_items, current_inventory)
            save_inventory(current_inventory)
        elif command == "show":
            print("Inventory:")
            if current_inventory:
                display_items(current_inventory)
            else:
                print("Empty")
        elif command == "drop":
            current_inventory = drop(current_inventory)
            save_inventory(current_inventory)
        elif command == "exit":
            save_inventory(current_inventory)
            print("Bye!")
            break
        else:
            print("Invalid command. Please try again.")
    print()

if __name__ == "__main__":
    main()
```
