```python
import random

ALL_ITEMS_FILE = "all_wizards_items.txt"
INVENTORY_FILE = "wizard_inventory.txt"

def load_all_items():
    with open(ALL_ITEMS_FILE, "r") as file:
        return [line.strip() for line in file]

def load_inventory():
    try:
        with open(INVENTORY_FILE, "r") as file:
            return [line.strip() for line in file]
    except FileNotFoundError:
        return []

def save_inventory(items):
    with open(INVENTORY_FILE, "w") as file:
        for item in items:
            file.write(item + "\n")
            
def display_items(items):
    for i in range(len(items)):
        print(str(i + 1) + ". " + items[i])

def walk(all_items, current_inventory):
    if len(current_inventory) >= 4:
        print("You can't carry any more items. Drop something first.")
        return current_inventory

    item_to_add = random.choice(all_items)
    print("While walking down a path, you see " + item_to_add + ".")
    grab = input("Do you want to grab it? (y/n): ").lower()
    if grab == "y":
        if len(current_inventory) >= 4:
            print("You can't carry any more items. Drop something first.")
        else:
            current_inventory.append(item_to_add)
            print("You picked up " + item_to_add + ".")
    return current_inventory

def drop(current_inventory):
    if not current_inventory:
        print("Inventory is empty.")
        return current_inventory

    print("Inventory:")
    display_items(current_inventory)
    try:
        index = int(input("Drop Number: ")) - 1
        if index < 0 or index >= len(current_inventory):
            print("Invalid item number.")
        else:
            dropped_item = current_inventory.pop(index)
            print("You dropped " + dropped_item +".")
    except ValueError:
        print("Invalid input. Please enter a number.")
    return current_inventory

def main():
    all_items = load_all_items()
    current_inventory = load_inventory()

    print("The Wizard Inventory Program")
    print()
    print("COMMAND MENU")
    print("walk - Walk down the path")
    print("show - Show all items")
    print("drop - Drop an item")
    print("exit - Exit program")

    while True:
        command = input("Command: ").lower()
        print()
        if command == "walk":
            current_inventory = walk(all_items, current_inventory)
            save_inventory(current_inventory)
        elif command == "show":
            print("Inventory:")
            if current_inventory:
                display_items(current_inventory)
            else:
                print("Empty")
        elif command == "drop":
            current_inventory = drop(current_inventory)
            save_inventory(current_inventory)
        elif command == "exit":
            save_inventory(current_inventory)
            print("Bye!")
            break
        else:
            print("Invalid command. Please try again.")

if __name__ == "__main__":
    main()
```

```python
import csv

MONTHS = {"Jan": 1, "Feb": 2, "Mar": 3, "Apr": 4, "May": 5, "Jun": 6,
    "Jul": 7, "Aug": 8, "Sep": 9, "Oct": 10, "Nov": 11, "Dec": 12}


MONTHS_NAMES = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]


def load_sales():
    sales = {}
    with open('monthly_sales.csv', 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            month, amount = row
            sales[month] = int(amount)
    return sales

def save_sales(sales):
    with open('monthly_sales.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        for month in MONTHS_NAMES:
            writer.writerow([month, sales[month]])

def display_monthly_sales(sales):
    print("Monthly Sales:")
    for month, amount in sales.items():
        print(month + " - " + str(amount))

def display_yearly_summary(sales):
    yearly_total = sum(sales.values())
    monthly_average = yearly_total / 12
    print("Yearly total:   ", yearly_total)
    print("Monthly average:   ", round(monthly_average, 2))

def edit_sales(sales):
    month = input("Three-letter Month: ")
    if month not in MONTHS:
        print("Invalid month abbreviation.")
        return

    amount = int(input("Sales Amount: "))
    sales[month] = amount
    save_sales(sales)
    print("Sales amount for " + month + " was modified.")

def main():
    sales = load_sales()

    print("Monthly Sales program")
    print()
    print("COMMAND MENU")
    print("monthly - View monthly sales")
    print("yearly  - View yearly summary")
    print("edit    - Edit sales for a month")
    print("exit    - Exit program")

    while True:
        print()
        command = input("Command: ")
        if command == "monthly":
            display_monthly_sales(sales)
        elif command == "yearly":
            display_yearly_summary(sales)
        elif command == "edit":
            edit_sales(sales)
        elif command == "exit":
            print("Bye!")
            break
        else:
            print("Invalid command. Please try again.")

if __name__ == "__main__":
    main()
```
