```python
import csv

def get_country():
    countries={}
    with open ("world_cup_champions.txt",  newline="") as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            country = row[1]
            if country in countries:
                countries[country]["Wins"] += 1
                countries[country]["Years"].append(row[0])
            else:
                countries[country] = {"Wins": 1, "Years": [row[0]]}
    return countries
            

def main(): 
    countries = get_country()
   
    print("Country\t       " + "Wins\t" + "Year")
    print("=======\t       " + "====\t" + "====\n")
 
    for country in sorted(countries.keys()):
        wins = countries[country]['Wins']
        years = ', '.join(countries[country]['Years'])
        print(f"{country.ljust(15)}{wins}\t{years}")

    
if __name__=="__main__":
    main()
    
```

```python
def view_sales(sales_data, month):
    month = month.lower()
    if month in sales_data:
        print(f"Sales amount for {month.capitalize()} is {sales_data[month]:,.2f}.")
    else:
        print("Invalid three-letter month.")
    print()
    
def load_sales_data(mothly_sales):
    sales_data = {}
    with open(mothly_sales, 'r') as file:
        for line in file:
            month, sales = line.strip().split('\t')
            sales_data[month.lower()] = float(sales)
    return sales_data

def save_sales_data(mothly_sales, sales_data):
    with open(mothly_sales, 'w') as file:
        for month, sales in sales_data.items():
            file.write(f"{month.capitalize()}\t{sales:.2f}\n")

def edit_sales(sales_data, month, new_amount):
    month = month.lower()
    if month in sales_data:
        sales_data[month] = new_amount
        print(f"Sales amount for {month.capitalize()} is {new_amount:,.2f}.")
    else:
        print("Invalid three-letter month.")
    print()    

def calculate_totals(sales_data):
    yearly_total = sum(sales_data.values())
    monthly_average = yearly_total / len(sales_data)
    print(f"Yearly total: {yearly_total:,.2f}")
    print(f"Monthly average: {monthly_average:,.2f}")
    print()

def main():
    mothly_sales = "monthly_sales.txt"
    sales_data = load_sales_data(mothly_sales)

    print("Monthly Sales program")
    print()
    print("COMMAND MENU")
    print("view\t- View sales for specified month")
    print("edit\t- Edit sales for specified month")
    print("totals\t- View sales summary for year")
    print("exit\t- Exit program")
    print()

    while True:
        command = input("Command: ").strip().lower()
        
        if command == "view":
            month = input("Three-letter Month: ").strip().lower()
            view_sales(sales_data, month)
        elif command == "edit":
            month = input("Three-letter Month: ").strip().lower()
            if month in sales_data:
                new_amount = float(input("Sales Amount: "))
                edit_sales(sales_data, month, new_amount)
                save_sales_data(mothly_sales, sales_data)
            else:
                print("Invalid three-letter month.")
        elif command == "totals":
            calculate_totals(sales_data)
        elif command == "exit":
            print("Bye!")
            break
        else:
            print("Invalid command. Please try again.")

if __name__ == "__main__":
    main()
```
